```r
# ggplot2 pruebas
library(ggplot2) 
```

```r
# crear factores con etiquetas de valor
mtcars$gear <- factor(mtcars$gear,levels=c(3,4,5),
  	labels=c("3gears","4gears","5gears")) 
mtcars$am <- factor(mtcars$am,levels=c(0,1),
  	labels=c("Automatic","Manual")) 
mtcars$cyl <- factor(mtcars$cyl,levels=c(4,6,8),
   labels=c("4cyl","6cyl","8cyl")) 
```

```r
# Gráficas de densidad del kernel para mpg
# agrupados por número de engranajes (indicados por color)
qplot(mpg, data=mtcars, geom="density", fill=gear, alpha=I(.5), 
   main="Distribución de Kilometraje de Gas", xlab="millas por galón", 
   ylab="Densidad")
```

```r
# Diagrama de dispersión de mpg vs. hp para cada combinación de engranajes y cilindros
# en cada faceta, el tipo de transmisión está representado por forma y color
qplot(hp, mpg, data=mtcars, shape=am, color=am, 
   facets=gear~cyl, size=I(3),
   xlab="Horsepower", ylab="Millas por Galon") 
```

```r
# Gráficos de caja de mpg por número de marchas
# observaciones (puntos) están superpuestas y alteradas
qplot(gear, mpg, data=mtcars, geom=c("boxplot", "jitter"), 
   fill=gear, main="Kilometraje por número de marcha",
   xlab="", ylab="Millas por galon")
```
